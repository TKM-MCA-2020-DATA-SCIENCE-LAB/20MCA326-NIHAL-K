```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,5,10)
y=x**2
x
```

```python
y
```

```python
plt.plot(x,y,'r')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('graph')
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r*-')
plt.subplot(2,2,2)
plt.plot(x,y,'--')
plt.subplot(2,2,3)
plt.plot(x,y,'b')
plt.show()
```

using object oriented


```python

fig=plt.figure()
axes=fig.add_axes([0.1,0.2,0.8,0.7])
axes.plot(x,y,'b')
axes.set_xlabel('x axis')
axes.set_ylabel('y axis')
axes.set_title('graph')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.4,0.2,0.8,0.7])
axes2=fig.add_axes([0.5,0.5,0.3,0.3])
axes1.plot(x,y,'b')
axes1.set_xlabel('x axis')
axes1.set_ylabel('y axis')
axes1.set_title('graph1')
axes2.plot(x,y,'r')
axes2.set_xlabel('x axis')
axes2.set_ylabel('y axis')
axes2.set_title('graph2')
plt.show()
```

subplots


```python
fig,axes=plt.subplots(1,2)
for a in axes:
  a.plot(x,y,'r')
  a.set_xlabel('x axis')
  a.set_ylabel('y axis')
  a.set_title("graph")
fig.tight_layout()
plt.show()
```

```python
fig, axes=plt.subplots(figsize=(5,4),dpi=100)
plt.plot(x,y,'*-')
plt.show()
```

LEGEND

```python
fig=plt.figure()
axes1=fig.add_axes([0.4,0.2,0.8,0.7])
axes1.plot(x,x**2,label="x**2")
axes1.plot(x,x**3,label="x**3")
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.legend(loc=2)
plt.show()
```

LINE COLOR,LINE WIDTH,LINE TYPE

```python
plt.plot(x,x+5,color='b',lw=3,ls='-',marker='o',markersize=12,linestyle='dashed',markeredgecolor='yellow',markeredgewidth=4,markerfacecolor='red')
plt.show()
```
